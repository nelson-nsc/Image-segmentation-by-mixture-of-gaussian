```python
import numpy as np
import matplotlib.pyplot as plt
import glob
%matplotlib inline
```

## Load Training Data

```python
def load_img(f_path):
    files = glob.glob(f_path)
    Imgs = []
    img_num = 1
    for f in files:
        img = plt.imread(f)
        Imgs.append(img)
        print("Image {}: {}".format(img_num, str(img.shape)))
        img_num += 1
    print("Imgs length:", len(Imgs))
    return Imgs
```

```python
data_color = load_img("apples/*.jpg")
data_bin = load_img("apples/*.png")
```

```python
def load_train(img_color, img_bin):
    data_true = []
    data_false = []
    #gtruth = (img_bin[:,:,2]).astype(int)
    for i in range(int(len(img_color))):
        for row in range(int(img_color[i].shape[0])):
            for col in range(int(img_color[i].shape[1])):
                ground_truth = np.array(img_bin[i])[:,:,2]
                if ground_truth[row, col] == 1:
                    data_true.append(img_color[i][row,col])
                else:
                    data_false.append(img_color[i][row,col])
    print('data_true: {} ; data_false: {}'.format(data_true,data_false))
    data_true = np.array(data_true)/255
    data_false = np.array(data_false)/255
    return data_true, data_false
```
